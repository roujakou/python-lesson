```python
# 處理錯誤
# 應用程式增加可靠度
try:
    money = int(input("請輸入金額"))
    if money < 0:
        raise Exception("金額不可以小於零")
except ValueError:
    print("格式錯誤")
except Exception as error:
    print("error")
else:
    print(money)
finally:
    print("應用程式結束")
```
